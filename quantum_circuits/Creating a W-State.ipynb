```python
from qiskit import *
from math import pi
from qiskit.visualization import plot_bloch_multivector, plot_state_city
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
circ = QuantumCircuit(3)
circ.ry(pi/3, 0)
circ.h(1)
circ.toffoli(0,1,2)
```

```python
out_state = execute(circ,backend).result().get_statevector() # Do the simulation, returning the state vector
```

```python
plot_bloch_multivector(out_state) # Display the output state vector
```

```python
plot_state_city(out_state)
```
